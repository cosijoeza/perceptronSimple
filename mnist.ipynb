# Perceptron de Rosenblott

**Solo 2 clases**

```python
x = np.concatenate((X[num1,:],X[num2,:]),axis = 0)
```

**Normalizar los datos**

```python
mms = MinMaxScaler()
XX = mms.fit_transform(x)
```

**Vector bias**

```python
b = np.ones(1000)
b = b.reshape(1000,1)
```

```python
#Concatenar bias a vector de caracteristicas
print(XX.shape)
print(b.shape)
XX = np.concatenate((b,XX),axis = 1)
print(XX)
```

**Etiquetas binarias. -1 => n; 1 => m**

¿Por qué ahora 1 y -1?
* Para distinguir los perceptrones
* Me he acostumbrado trabajarlos así, solo que el perceptron de mxlxtend solo permite 1y 0

```python
YY = np.arange(0,1000)
YY = (YY >= 500) * 2 -1
print(YY.shape)
print(YY)
```

```python
#Vector de pesos
W = np.random.uniform(low = -0.2,high = 0.2,size = 785) #Contando el bias
W.shape
```

### Definir la función signo

```python
def signo(X):
    return (X>=0) * 2 -1
```

**Separar los datos en entrenamiento y en test**

```python
test_size = 0.33
X_train,X_test,Y_train,Y_test = train_test_split(XX,YY,test_size=0.33)
```

### Ejecutar el algoritmo del perceptron

```python
factor = 0.2
niters = 100
history = np.array([W])
rendimientos = np.array([])
for i in range(niters):
    #Numero aleatorio
    j = np.random.randint(0,1000-(1000*test_size),1)
    #Multiplicación
        #----------------v = XX[j,:].dot(W.T)
    v = X_train[j,:].dot(W.T)   
    #Función de activación, valor obtenido
    y = signo(v)
    #Actualización de los datos
        #----------------W = W + factor *(YY[j]-y) * XX[j,:]
    W = W + factor * (Y_train[j]-y) * X_train[j,:]
    #Guardar vector de pesos
    history = np.append(history,W,axis = 0)
    
    #Calcular el rendimiento del peso actual
        #----------------respuesta = signo(XX.dot(W.T).reshape(1000))
        #----------------performance = (respuesta == YY).mean()
    respuesta = signo(X_train.dot(W.T).reshape(670))
    performance = (respuesta == Y_train).mean()
        
    #Guardar rendimientos
    rendimientos = np.append(rendimientos,performance)
    if(i > 0 and performance > rendimientos[i-1]):
        best_W = W
        best_pfm = performance
        best_i = i
```

**Dimensión del mejor W, indice del mejor W y precisión de mejor clasificación**

```python
print("Dimensión: ",best_W.shape,"\nIteración: ",best_i,"\nMejor rendimiento: ",best_pfm)
```

```python
#Obtener vector de respuestas generadas por el mejor modelo
    #resp = signo(XX.dot(best_W.T))
resp = signo(X_test.dot(best_W.T))
resp = resp.reshape(330)
```

**Cómo fue el cambio de la precisión en la clasificación**

```python
plt.plot(np.arange(niters),rendimientos)
```

# Resultados de la clasificación

```python
plt.rcParams["figure.figsize"] = (20,12)
plt.subplots_adjust()

for i in np.arange(0,50):
    indx = np.random.randint(0,1000*test_size)
    img = X_test[indx][1:].reshape(28,28) # [1:] es para eliminar una columna
    plt.subplot(5,10,i+1)
    plt.title("Etq: " + str(n if resp[indx] < 0 else m))
    plt.imshow(img)
```
